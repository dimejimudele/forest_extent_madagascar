This script plots the agreement proportions for the different years

```python
import numpy as np
import matplotlib.pyplot as plt
import proplot as pplt
```

```python
plt.rcParams.update({
    'xtick.labelsize': 20,
    'ytick.labelsize': 20,
    'legend.fontsize': 23,
    'axes.labelsize': 24,
    'axes.titlesize': 24
})
```

```python
years = [2016, 2017, 2018, 2019, 2020]
categories = ['Low', 'Medium', 'High']
data_all = np.array([
    [0.4762686631, 0.1952759436, 0.3284553933],  # 2016
    [0.3833339066, 0.2892538587, 0.3274122347],  # 2017
    [0.4345898819, 0.3606897487, 0.2047203694],  # 2018
    [0.4371569394, 0.3599488802, 0.2028941805],  # 2019
    [0.4400152115, 0.3386557274, 0.2213290611]   # 2020
])
```

```python
data_all = data_all * 100 #Convert to percent
```

```python
# Stacked Bar Plot with simplified legend
fig, ax = pplt.subplots(figsize=(8, 6))
colors = ['red', 'yellow', 'green']
bottom = np.zeros(len(years))

for i, (category, color) in enumerate(zip(categories, colors)):
    ax.barh(years, data_all[:, i], left=bottom, label=category, color=color)
    bottom += data_all[:, i]

ax.format(
    xlabel='Proportion (%)',
    ylabel='Year',
    xticklabelsize=18,
    yticklabelsize=18,
    xlabelsize=18,
    ylabelsize=18,
    titlesize=21,
)
ax.legend(bbox_to_anchor=(0, 1.15), loc='upper left', fontsize=30)

plt.savefig("agreement_proportions_whole_madagascar.pdf", format='pdf', dpi=300)
plt.show()

```

```python
biomes = ['Spiny', 'Dry', 'Moist', 'Mangroves']
categories = ['Low', 'Medium', 'High']

# Plotting the data using proplot
fig, ax = pplt.subplots(figsize=(8, 6))
colors = ['red', 'yellow', 'green']
bottom = np.zeros(len(biomes))

for i, (category, color) in enumerate(zip(categories, colors)):
    ax.barh(biomes, combined_data[:, i], left=bottom, label=category, color=color)
    bottom += combined_data[:, i]

ax.format(
    xlabel='Proportion (%)',
    ylabel='Biome',
    xticklabelsize=18,
    yticklabelsize=18,
    xlabelsize=18,
    ylabelsize=18,
    titlesize=21,
)
ax.legend(bbox_to_anchor=(0, 1.15), loc='upper left', fontsize=30)

plt.savefig("agreement_proportions_mean_biomes.pdf", format='pdf', dpi=300)
plt.show()
```
